```python
img=cv2.imread(r"C:\Users\acer\OneDrive\Desktop\Pics\resizedimg.png")
cv2.imshow('OG',img)
cv2.waitKey(0) # keeps the image displayed unti a keypress to close it

#grayscale img
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#blur img
blur_img=cv2.GaussianBlur(gray_img,(3,3),0)
cv2.imshow('Blur-Gray',blur_img)
cv2.waitKey(0)

#Using Canny method

# a function to automatically tune and find the best thresholds for the canny function: 
# credits to https://pyimagesearch.com/2015/04/06/zero-parameter-automatic-canny-edge-detection-with-python-and-opencv/

def auto_canny(image,sigma=0.33):
    med=np.median(image)

    lower=int(max(0,(1-sigma)*med))
    upper=int(min(255,(1+sigma)*med))

    edges=cv2.Canny(image,lower,upper)

    return edges

edge_img=cv2.Canny(blur_img,50,120)
cv2.imshow('Edges',edge_img)
pixels=np.array(list(edge_img))
cv2.waitKey(0)
#print(pixels)
save_img=Image.fromarray(pixels.astype(np.uint8))
save_img.save(r"C:\Users\acer\OneDrive\Desktop\Pics\Canny_img.jpeg",'JPEG')
```
